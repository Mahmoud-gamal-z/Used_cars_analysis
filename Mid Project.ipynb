```python
df['Car Brand'] = df['Car Model'].apply(lambda i:"".join(i.split(" ")[0]))
df['Car_Model_Name'] = df['Car Model'].apply(lambda i:" ".join(i.split(" ")[1:-1]))
df['Car_Model_Year'] = df['Car Model'].apply(lambda i:"".join(i.split(" ")[-1]))
```

```python
df.sample(3)
```

```python
df[['Average price','Minimum price','Maximum price']].sample(3)
```

```python
def change_price_to_numeric(r):
    return float(("".join(r.split(" ")[0])).replace(",",""))
```

```python
df['Average price'] = df['Average price'].astype(str).apply(change_price_to_numeric)
df['Minimum price'] = df['Minimum price'].astype(str).apply(change_price_to_numeric)
df['Maximum price'] = df['Maximum price'].astype(str).apply(change_price_to_numeric)
```

```python
df[['Average price','Minimum price','Maximum price']].sample(3)
```

```python
df['Month/Year'].sample(3)
```

```python
df['Month/Year'] = pd.to_datetime(df['Month/Year'],errors="coerce")
df['Month'] = df['Month/Year'].dt.month_name()
df['Quarter'] = df['Month/Year'].dt.quarter.astype(str).apply(lambda r:r.split(".")[0])
df['Year'] = df['Month/Year'].dt.year.astype(str).apply(lambda r:r.split(".")[0])
```
